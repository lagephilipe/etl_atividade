```python
# Philipe Lage
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
import MySQLdb
```

```python
#Extraindo dados do arquivo address.csv
csv_address = "bases/address.csv"
address_df = pd.read_csv(csv_address)
address_df.head()
```

```python
#Extraindo dados do arquivo people.csv
csv_people = "bases/people.csv"
people_df = pd.read_csv(csv_people, sep=','  , engine='python')
people_df.head()
```

```python
# Extraindo dados do arquivo cars.json
import json
from pandas.io.json import json_normalize
cat=[]
```

```python
json_file="bases/cars.json"
type(json_file)
```

```python
df_cars_json=pd.read_json(json_file)
df_cars_json.head()
```

```python
#Unindo os 3 Data frames em um só
final_df=pd.concat([people_df,address_df,df_cars_json], axis=1, join='inner')
final_df = final_df.reset_index(drop=True)
final_df.insert(0,'ID',range(1,1+len(final_df)))
final_df=final_df.set_index('ID')

final_df
```

```python
# Ajustando colunas
final_df2 = final_df.drop(columns=['id_people'])
final_df2
```

```python
#final_df2 = final_df2.rename(columns={'ID': 'id_people'})
final_df2.insert(0,'id_people',range(1,1+len(final_df)))
final_df2=final_df2.set_index('id_people')
final_df2
```

```python
#Salvando resultado em csv
final_df2.to_csv('saida_dados/people_final.csv')
```

```python
#CARGA DE DADOS NO MYSQL

import csv
import pymysql
from config import config
 
cnx = pymysql.connect(host='localhost',
                         user='root',
                         password='root',
                         db='atividadegotodata', charset='utf8')
cursor = cnx.cursor()
 
input_file = csv.DictReader(open('output_data/people_final.csv', encoding='utf-8'))
for row in input_file:
    cursor.execute("INSERT INTO atividadegotodata.people (id_people ,first_name , last_name , email , gender , language , company_name ,department , country_code , country , city , street_address , car_model , car_model_year) \
                    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s)",(row['id_people'],row['first_name'],row['last_name'],row['email'],
                                                                                     row['gender'],row['language'],row['company_name'],row['department'],
                                                                                     row['country_code'],row['country'],row['city'],row['street_address'],
                                                                                     row['car_model'],row['car_model_year']))
    cnx.commit()
    print('Num: {}, People: {}'.format(row['id_people'], row['first_name']))

```
